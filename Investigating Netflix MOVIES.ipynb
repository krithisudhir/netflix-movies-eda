```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('netflix_data.csv')
df.head()
```

```python
df.info()

```

```python
print(df['type'].unique())
print(df['genre'].unique())
print(df['country'].unique())
```

```python
#No data quality issues
```

```python
#subset data to only study movies
netflix_subset=df[df['type']=='Movie']
netflix_subset.head()
```

```python
print(df.size)
print(netflix_subset.size)
```

```python
#keep only relavant columns
netflix_movies=netflix_subset[['title','country','genre','release_year','duration']]
netflix_movies.head()
short_movies=netflix_movies[netflix_movies['duration']<60]
short_movies.head(20)
```

```python
#perform a preliminary analysis on variables to see what could potentially be influencing the duration of movies
dur_by_country=short_movies[['country','duration']]
dur_by_genre=short_movies[['genre','duration']]
dur_by_year=short_movies[['release_year','duration']]
```

```python

import seaborn as sns

#year vs duration
sns.boxplot(x='release_year', y='duration', data=dur_by_year)
plt.title('Duration by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Duration')
plt.xticks(rotation=90)
plt.show()
#it's getting shorter since 2017
```

```python
#genre vs duration
median1 = dur_by_genre.groupby('genre')['duration'].median()
# Sort categories based on median values in descending order
desired_order = median1.sort_values(ascending=True).index
sns.boxplot(x='genre', y='duration', data=dur_by_genre,order=desired_order)
plt.title('Duration by genre')
plt.xlabel('Genre')
plt.ylabel('Duration')
plt.xticks(rotation=90)
plt.show()
#of all the short films, Dramas, children , horror and classic have the least times
```

```python
median_all = netflix_movies.groupby('genre')['duration'].median()
# Sort categories based on median values in descending order
desired_order = median_all.sort_values(ascending=True).index
sns.boxplot(x='genre', y='duration', data=netflix_movies,order=desired_order)
plt.title('Duration by genre')
plt.xlabel('Genre')
plt.ylabel('Duration')
plt.xticks(rotation=90)
plt.show()
```

```python
#duration by genre trend is different when compared to entire dataset
```

```python
#country vs duration
median2 = dur_by_country.groupby('country')['duration'].median()
# Sort categories based on median values in descending order
desired_order = median2.sort_values(ascending=True).index
sns.boxplot(x='country', y='duration', data=dur_by_country,order=desired_order)
plt.title('Duration by country')
plt.xlabel('country')
plt.ylabel('Duration')
plt.xticks(rotation=90)
plt.show()

#Aus, Canada and Netherlands for sure have shorter films and have more data points
```

```python

#observe similar plots for entire dataset
#prepare color code for different genres to plot
def final_category(x):
    if x=="Children":
        return x
    elif x=="Documentaries":
        return x
    elif x=="Stand-Up" :
        return x
    else:
        return "Other"
# netflix_movies['colors']=netflix_movies['genre'].apply(lambda x : color_code(x))
# netflix_movies.head()
netflix_movies['category_final']=netflix_movies['genre'].apply(lambda x : final_category(x))
netflix_movies.head()

custom_palette = {"Children": "orange", "Other": "gray", "Documentaries": "green","Stand-Up":"black"} 
sns.scatterplot(x='release_year', y='duration', data=netflix_movies, hue='category_final',palette=custom_palette)
plt.title('Movie Duration by Year of Release')
plt.ylabel("Duration(min)")
plt.xlabel('Release Year')
plt.legend(title='Genre')
plt.show()
```

```python
#while some genres tend to have shorter duration, we cannot say anything about the duration of all releases in general
#documentaries and CHILDREN'S content have increased around the 2000s, so a regular viewers of such content may feel the duration has been reducing
```
